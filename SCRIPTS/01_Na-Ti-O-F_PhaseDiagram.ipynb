# TiOF$_2$

```python
PD_1 = CompoundPhaseDiagram(entries,
                            terminal_compositions=[Composition("O2"),
                                                   Composition("TiF3"),
                                                   Composition("TiO2")],
                           normalize_terminal_compositions=False)

compound_plotter_1  = PDPlotter(PD_1, show_unstable=True, backend = "matplotlib",  markersize=20)
z1                  = compound_plotter_1.get_plot(label_unstable=False, label_stable=True)

z1.show()
```

# NaTiOF$_2$
